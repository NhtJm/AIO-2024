##**Cosine Similarity**

```python
import numpy as np

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("A:", A)
print("B:", B)

# compute cosine similarity
cosine = np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))
print("Cosine Similarity:", cosine)
```

###**Image Similarity**

```python
from PIL import Image

image1 = Image.open('./data/image1.jpg')
image2 = Image.open('./data/image2.jpg')
query_image = Image.open('./data/query_image.jpg')
```

```python
image1_data = np.array(image1)
image2_data = np.array(image2)
query_image_data = np.array(query_image)
image1_data.shape, image2_data.shape, query_image_data.shape
```

```python
image1_data = np.sum(image1_data, axis=-1, keepdims=1)/3
image2_data = np.sum(image2_data, axis=-1, keepdims=1)/3
query_image_data = np.sum(query_image_data, axis=-1, keepdims=1)/3
```

```python
image1_flatten = image1_data.flatten()
image2_flatten = image2_data.flatten()
query_image_flatten = query_image_data.flatten()
```

```python
def compute_cosine_similarity(A, B):
    cosine = np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))
    return cosine
```

```python
cosine1 = compute_cosine_similarity(
    image1_flatten, query_image_flatten
)
cosine2 = compute_cosine_similarity(
    image2_flatten, query_image_flatten
)
cosine1, cosine2
```
